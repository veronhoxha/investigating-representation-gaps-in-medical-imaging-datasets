```python


def resize_images_in_folder(input_folder, output_folder, new_size):
    """
    Resize all images in the input_folder and save the resized images to the output_folder.

    Parameters:
    input_folder (str): Path to the folder containing input images.
    output_folder (str): Path to the folder where resized images will be saved.
    new_size (tuple): New size of the images, e.g., (width, height).
    """
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Iterate over each file in the input folder
    for filename in os.listdir(input_folder):
        if filename.endswith((".jpg", ".png")):  # Check for image file extensions
            # Open the image
            with Image.open(os.path.join(input_folder, filename)) as img:
                # Resize the image
                resized_img = img.resize(new_size)
                # Save the resized image to the output folder
                img.show()
                resized_img.show()
                resized_img.save(os.path.join(output_folder, filename))

# Example usage:
input_folder = "../data/HAM10000/HAM10000_images_part_1"
output_folder = "../data/data_resized/"
new_size = (100, 100)  # Specify the new size here, e.g., (width, height)
resize_images_in_folder(input_folder, output_folder, new_size)
```

```python


def adjust_brightness_in_folder(input_folder, output_folder, brightness_factor):
    """
    Adjust the brightness of all images in the input_folder and save the adjusted images to the output_folder.

    Parameters:
    input_folder (str): Path to the folder containing input images.
    output_folder (str): Path to the folder where adjusted images will be saved.
    brightness_factor (float): Factor to adjust the brightness.
                              1.0 means no change, less than 1.0 darkens the image, and greater than 1.0 brightens the image.
    """
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Iterate over each file in the input folder
    for filename in os.listdir(input_folder):
        if filename.endswith((".jpg", ".png")):  # Check for image file extensions
            # Open the image
            with Image.open(os.path.join(input_folder, filename)) as img:
                # Adjust the brightness
                enhancer = ImageEnhance.Brightness(img)
                adjusted_img = enhancer.enhance(brightness_factor)
                # Save the adjusted image to the output folder
                adjusted_img.save(os.path.join(output_folder, filename))
               

# Example usage:
input_folder = "../data/HAM10000/HAM10000_images_part_1"
input_folder2 = "../data/HAM1000_2"
output_folder = "../data/data_bright/"
brightness_factor = .7  # Adjust the brightness factor as needed
adjust_brightness_in_folder(input_folder, output_folder, brightness_factor)
#adjust_brightness_in_folder(input_folder2, output_folder, brightness_factor)
```
